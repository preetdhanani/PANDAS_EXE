### Step 12.  revenue from each item 



```python
chipo['total_bill'] = chipo.quantity * chipo.item_price
```

```python
sns.set(rc={"figure.figsize":(20, 10)})
chipo.groupby('item_name').total_bill.sum().plot.bar()
```

### Step 13. max quantity  per order

```python
chipo.groupby('order_id').quantity.sum()
```

```python
chipo.groupby('order_id').quantity.sum().sort_values(ascending=False)
```

```python
max_quantity_orderid = chipo.groupby('order_id').quantity.sum().argmax() + 1 # index start with 1 in dataset
print(max_quantity_orderid)
chipo[chipo.order_id == max_quantity_orderid]
```

### Step 13.  which item has sold max

```python
ax =chipo.groupby('item_name').quantity.sum().plot.bar()
ax.bar_label(ax.containers[0]);
```
